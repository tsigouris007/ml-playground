## Matplotlib example workflow

```python
# 0. import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# 1. prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]
```

```python
# 2. setup plot
fig, ax = plt.subplots(figsize=(10, 10)) # figsize(width, height)

# 3. plot data
ax.plot(x, y)

# 4. customize plot
ax.set(
    title="Simple plot",
    xlabel="x-axis",
    ylabel="y-axis",
);

# 5. save & show
fig.savefig("./images/sample-plot.png")
```

### Two options for subplots

```python
# Option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(
    nrows=2,
    ncols=2,
    figsize=(10, 5)
)
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(butter_prices.keys(), butter_prices.values())
ax4.hist(np.random.randn(1000));
```

```python
# Option 2
fig, ax =plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

# Plot for each index
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].barh(list(butter_prices.keys()), list(butter_prices.values()))
ax[1, 1].hist(np.random.randn(1000));
```

## pyplot vs matplotlib OO method
- When plotting something quickly, use pyplot
- When plotting something advanced, use OO method

```python
over_50 = heart_disease[heart_disease.age > 50]
over_50
```

```python
# pyplot
over_50.plot(kind="scatter", x="age", y="chol", c="target", figsize=(10, 6));
```

```python
## matplotlib OO
fix, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter", x="age", y="chol", c="target", ax=ax)
ax.set_xlim([45, 80]); # sets the x axis limits
```

```python
# 00 from scratch
fig, ax = plt.subplots(figsize=(10, 6))
# Plot data
scatter = ax.scatter(x=over_50.age, y=over_50.chol, c=over_50.target)
# Customize
ax.set(
    title="Heart Disease & Cholesterol Levels",
    xlabel="Age",
    ylabel="Cholesterol"
);
# Legend
ax.legend(*scatter.legend_elements(), title="Target")
# Add horizontal line
ax.axhline(y=over_50.chol.mean(), linestyle="dotted");
```

```python
# 00 from scratch
fig, ax = plt.subplots(figsize=(10, 6))
# Plot data
scatter = ax.scatter(x=over_50.age, y=over_50.thalach, c=over_50.target)
# Customize
ax.set(
    title="Heart Disease & Max Heart Rate Levels",
    xlabel="Age",
    ylabel="Max Heart Rate"
);
# Legend
ax.legend(*scatter.legend_elements(), title="Target")
# Add horizontal line
ax.axhline(y=over_50.thalach.mean(), linestyle="dotted")
# Set x axis limits
ax.set_xlim([45, 80]);
```

## Extra plots
### Age, Cholesterol, Max Heart Rate

```python
# 00 from scratch
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10, 10), sharex=True)
# Add data ax0
scatter = ax0.scatter(x=over_50.age, y=over_50.chol, c=over_50.target)
ax0.set(
    title="Heart Disease & Cholesterol Levels",
    # xlabel="Age",
    ylabel="Cholesterol"
)
# legend for ax0
ax0.legend(*scatter.legend_elements(), title="Target")
# Horizontal line for ax0
ax0.axhline(y=over_50.chol.mean(), linestyle="dotted")
# Add data ax1
scatter = ax1.scatter(x=over_50.age, y=over_50.thalach, c=over_50.target)
ax1.set(
    title="Heart Disease & Max Heart Rate Levels",
    xlabel="Age",
    ylabel="Max Heart Rate"
)
# Legend for ax1
ax1.legend(*scatter.legend_elements(), title="Target")
# Horizontal line for ax1
ax1.axhline(y=over_50.thalach.mean(), linestyle="dotted")
# Add figure title
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight=700);
```

```python
# Set style
# plt.style.use('seaborn-v0_8-whitegrid')
plt.style.use('dark_background')

# 00 from scratch
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10, 10), sharex=True)
# Add data ax0
scatter = ax0.scatter(
    x=over_50.age,
    y=over_50.chol,
    c=over_50.target,
    cmap="winter"
)
ax0.set(
    title="Heart Disease & Cholesterol Levels",
    # xlabel="Age",
    ylabel="Cholesterol"
)
# legend for ax0
ax0.legend(*scatter.legend_elements(), title="Target")
# Horizontal line for ax0
ax0.axhline(y=over_50.chol.mean(), linestyle="dotted", c="red")
# Set x, y axis limit for ax0
ax0.set_xlim([50, 80])
ax0.set_ylim([100, 600])

# Add data ax1
scatter = ax1.scatter(
    x=over_50.age,
    y=over_50.thalach,
    c=over_50.target,
    cmap="summer")
ax1.set(
    title="Heart Disease & Max Heart Rate Levels",
    xlabel="Age",
    ylabel="Max Heart Rate"
)
# Legend for ax1
ax1.legend(*scatter.legend_elements(), title="Target")
# Horizontal line for ax1
ax1.axhline(y=over_50.thalach.mean(), linestyle="dotted", c="red")
# Add figure title
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight=700)
# Set y axis limit for ax1
ax1.set_ylim([60, 200]);
```
